```python
print("======================================================")
print("        IA PARA ANÁLISIS DE LÍNEA DE CRÉDITO")
print("======================================================")

nombre = input("Ingrese su nombre: ")

while True:

    print("\n=================== MENÚ ===================")
    print("1. Solicitar línea de crédito")
    print("2. Consultar estado financiero")
    print("3. Simular préstamo")
    print("4. Salir")
    print("============================================")

    opcion = input("Seleccione una opción: ")

    # ==================================================
    # OPCIÓN 1
    # ==================================================

    if opcion == "1":

        print("\n========== SOLICITUD DE CRÉDITO ==========")

        edad = int(input("Ingrese su edad: "))
        salario = float(input("Ingrese su salario mensual: "))
        antiguedad = int(input("Ingrese años trabajando: "))

        if edad >= 18 and salario >= 8000:

            print("\nLa IA detectó que cumple requisitos básicos.")

            if antiguedad >= 2:

                print("Resultado:")
                print("- Crédito aprobado")
                print("- Línea recomendada: $50,000")
                print("- Riesgo financiero bajo")

            elif antiguedad == 1:

                print("Resultado:")
                print("- Crédito aprobado parcialmente")
                print("- Línea recomendada: $20,000")
                print("- Riesgo financiero medio")

            else:

                print("Resultado:")
                print("- Crédito limitado")
                print("- Línea recomendada: $10,000")
                print("- Riesgo financiero alto")

        else:

            print("\nLa IA rechazó la solicitud.")
            print("Motivos posibles:")
            print("- Edad insuficiente")
            print("- Salario muy bajo")

    # ==================================================
    # OPCIÓN 2
    # ==================================================

    elif opcion == "2":

        print("\n========== ESTADO FINANCIERO ==========")

        ingresos = float(input("Ingrese sus ingresos mensuales: "))
        gastos = float(input("Ingrese sus gastos mensuales: "))

        if ingresos > gastos:

            ahorro = ingresos - gastos

            print("\nLa IA analizó sus finanzas.")
            print("Ahorro mensual:", ahorro)

            if ahorro >= 10000:

                print("Estado financiero: Excelente")
                print("Posibilidad alta de nuevos créditos")

            elif ahorro >= 5000:

                print("Estado financiero: Bueno")
                print("Posibilidad media de nuevos créditos")

            else:

                print("Estado financiero: Regular")
                print("Se recomienda reducir gastos")

        else:

            print("\nLa IA detectó problemas financieros.")
            print("Sus gastos superan sus ingresos.")

    # ==================================================
    # OPCIÓN 3
    # ==================================================

    elif opcion == "3":

        print("\n========== SIMULADOR DE PRÉSTAMO ==========")

        monto = float(input("Ingrese el monto deseado: "))
        plazo = int(input("Ingrese el plazo en meses: "))

        if monto <= 10000:

            interes = monto * 0.10

            print("\nLa IA calculó un interés del 10%")

        elif monto <= 50000:

            interes = monto * 0.18

            print("\nLa IA calculó un interés del 18%")

        else:

            interes = monto * 0.25

            print("\nLa IA calculó un interés del 25%")

        total = monto + interes
        mensualidad = total / plazo

        print("\n=========== RESULTADO ===========")
        print("Monto solicitado:", monto)
        print("Interés generado:", interes)
        print("Total a pagar:", total)
        print("Pago mensual:", mensualidad)

        if mensualidad > 15000:

            print("La IA considera que el préstamo es riesgoso.")

        else:

            print("La IA considera que el préstamo es accesible.")

    # ==================================================
    # SALIR
    # ==================================================

    elif opcion == "4":

        print("\nPROGRAMA FINALIZADO")
        break

    # ==================================================
    # ERROR
    # ==================================================

    else:

        print("\nOpción incorrecta")
print("Programa hecho por Jaquez Andres 0074")
```
